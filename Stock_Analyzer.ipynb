## Step 4: Read Your Report
Run this cell after the analysis finishes to see the full report.

```python
import glob
from IPython.display import display, Markdown

# Find the most recent output folder
output_dirs = sorted(glob.glob('output/*/'), reverse=True)

if not output_dirs:
    print('❌ No output found yet. Make sure Step 3 finished running.')
else:
    latest = output_dirs[0]
    report_path = f'{latest}final_report.md'
    
    if os.path.exists(report_path):
        print(f'📄 Report from: {latest}')
        print('=' * 70)
        with open(report_path, 'r') as f:
            report = f.read()
        display(Markdown(report))
    else:
        print(f'⏳ Analysis folder exists ({latest}) but final report not generated yet.')
        print('The analysis may still be running or may have encountered an error.')
        print(f'\nFiles so far:')
        for f in sorted(os.listdir(latest)):
            print(f'  {f}')
```

---
## Step 5 (Optional): Read Individual Agent Reports
Drill into any specific agent's full analysis.

```python
import glob, os
from IPython.display import display, Markdown

output_dirs = sorted(glob.glob('output/*/'), reverse=True)
if not output_dirs:
    print('No output found.')
else:
    latest = output_dirs[0]
    files = sorted([f for f in os.listdir(latest) if f.endswith('_full.md')])
    
    if not files:
        print('No agent reports found yet.')
    else:
        print('Available agent reports:')
        for i, f in enumerate(files):
            print(f'  [{i}] {f}')
        
        choice = input('\nEnter a number to read (or "all" to see everything): ').strip()
        
        if choice.lower() == 'all':
            for f in files:
                with open(f'{latest}{f}', 'r') as fh:
                    display(Markdown(f'\n---\n## {f}\n'))
                    display(Markdown(fh.read()))
        elif choice.isdigit() and 0 <= int(choice) < len(files):
            with open(f'{latest}{files[int(choice)]}', 'r') as fh:
                display(Markdown(fh.read()))
        else:
            print('Invalid choice.')
```

## Step 6 (Optional): Download Report Files
Download the full report and raw data to your computer.

```python
import glob, shutil
from google.colab import files as colab_files

output_dirs = sorted(glob.glob('output/*/'), reverse=True)
if not output_dirs:
    print('No output found.')
else:
    latest = output_dirs[0]
    zip_name = latest.rstrip('/').split('/')[-1]
    shutil.make_archive(f'/content/{zip_name}', 'zip', latest)
    print(f'📦 Downloading {zip_name}.zip...')
    colab_files.download(f'/content/{zip_name}.zip')
```
